```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("temperature_data.csv")
```

```python
ddf=df.head(10)
print(ddf)
```

```python
ddf["Date Time"]=pd.to_datetime(ddf["Date Time"])
```

```python
x=ddf["Date Time"]
```

```python
y=ddf["T (degC)"]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x,y)
plt.xlabel("DATETIME")
plt.ylabel("TEMPERATURE IN DEGREES CELCIUS")
plt.title("TEMPERATURE DATA")
plt.grid()
plt.scatter(x,y)
```
